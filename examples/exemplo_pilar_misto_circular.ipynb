# Exemplo: pilar misto circular preenchido

Este notebook mostra uma criação simples de um `PilarCircularPreenchido` e imprime algumas características principais da seção: áreas, capacidade axial e momentos resistentes à flexão.

As unidades seguem o padrão usado no pacote: dimensões em `mm`, tensões em `MPa`, forças em `N` e momentos em `N.mm`.

## 1. Importação das classes

Este exemplo assume que a biblioteca `MY_PACKAGE` já está instalada no ambiente Python usado pelo notebook.

```python
from MY_PACKAGE.domain.value_objects.pilar_misto_circular import PilarCircularPreenchido
from MY_PACKAGE.domain.value_objects.ObjetoConcreto import ConcretoNormal
from MY_PACKAGE.domain.value_objects.ObjetoAco import AcoEstrutural, AcoArmadura
```

## 2. Definição dos materiais

Neste exemplo usamos concreto normal de `fck = 40 MPa`, aço estrutural de `fy = 345 MPa` e armadura de `fy = 500 MPa`.

```python
concreto = ConcretoNormal(fck=40, modulo_elasticidade=30000)
aco_estrutural = AcoEstrutural(fy=345)
aco_armadura = AcoArmadura(fy=500)

print(f"Concreto: fck = {concreto.fck:.1f} MPa | fcd = {concreto.fcd:.2f} MPa")
print(f"Aço estrutural: fy = {aco_estrutural.fy:.1f} MPa | fyd = {aco_estrutural.resistencia_design:.2f} MPa")
print(f"Armadura: fy = {aco_armadura.fy:.1f} MPa | fsd = {aco_armadura.resistencia_design:.2f} MPa")
```

## 3. Criação do pilar misto circular

A seção abaixo cria um tubo circular preenchido com concreto e com armadura longitudinal.

```python
pilar = PilarCircularPreenchido(
    diametro_tubo=600,
    espessura_tubo=9.5,
    material_aco_estrutural=aco_estrutural,
    material_concreto=concreto,
    material_armadura=aco_armadura,
    diametro_armadura_longitudinal=20,
    numero_armadura_longitudinal=10,
    diametro_armadura_transversal=5,
    espacamento_armadura_transversal=150,
    cobrimento=25,
    comprimento_pilar_destravado=3000,
)

print("Pilar criado com sucesso!")
print(f"Diâmetro externo do tubo: {pilar.diametro_tubo:.1f} mm")
print(f"Espessura do tubo: {pilar.espessura_tubo:.1f} mm")
print(f"Diâmetro interno: {pilar.diametro_interno:.1f} mm")
print(f"Comprimento destravado: {pilar.comprimento_pilar_destravado:.1f} mm")
```

## 4. Características geométricas

A área total é calculada pela API pública `area_total()`, enquanto as parcelas de aço, concreto e armadura são propriedades do pilar.

```python
print("Características geométricas")
print(f"Área de aço do tubo: {pilar.area_aco:,.2f} mm²")
print(f"Área de concreto: {pilar.area_concreto:,.2f} mm²")
print(f"Área de armadura: {pilar.area_armadura:,.2f} mm²")
print(f"Área total: {pilar.area_total():,.2f} mm²")
```

## 5. Capacidade axial

Nesta célula imprimimos as parcelas de capacidade axial plástica de cálculo e a capacidade axial resistente do pilar considerando o fator de redução por esbeltez.

```python
print("Capacidade axial de cálculo")
print(f"Npl,a,Rd: {pilar.capacidade_axial_plastico_aco_design() / 1_000:,.2f} kN")
print(f"Npl,c,Rd: {pilar.capacidade_axial_plastico_concreto_design() / 1_000:,.2f} kN")
print(f"Npl,s,Rd: {pilar.capacidade_axial_plastico_armadura_design() / 1_000:,.2f} kN")
print(f"Npl,Rd: {pilar.capacidade_axial_plastico_design() / 1_000:,.2f} kN")
print(f"Fator de redução: {pilar.fator_reducao:.4f}")
print(f"NRd do pilar: {pilar.capacidade_axial_resistente_pilar_design / 1_000:,.2f} kN")
```

## 6. Capacidade à flexão

Para este perfil circular, os momentos nos eixos `xx` e `yy` ficam muito próximos. O exemplo imprime o momento plástico resistente de cálculo e o momento plástico máximo de cálculo.

```python
print("Capacidade à flexão de cálculo")
print(f"Mpl,Rd,xx: {pilar.momento_resistente_plastico_total_design_xx / 1_000_000:,.2f} kN.m")
print(f"Mmax,Rd,xx: {pilar.momento_resistente_maximo_plastico_total_design_xx / 1_000_000:,.2f} kN.m")
print(f"Mpl,Rd,yy: {pilar.momento_resistente_plastico_total_design_yy / 1_000_000:,.2f} kN.m")
print(f"Mmax,Rd,yy: {pilar.momento_resistente_maximo_plastico_total_design_yy / 1_000_000:,.2f} kN.m")
```

## 7. Resumo rápido

A célula final reúne os principais resultados em uma estrutura simples, útil para copiar para relatórios ou comparar diferentes seções.

```python
resumo = {
    "area_total_mm2": pilar.area_total(),
    "capacidade_axial_plastica_design_kN": pilar.capacidade_axial_plastico_design() / 1_000,
    "capacidade_axial_resistente_pilar_design_kN": pilar.capacidade_axial_resistente_pilar_design / 1_000,
    "momento_plastico_design_xx_kNm": pilar.momento_resistente_plastico_total_design_xx / 1_000_000,
    "momento_plastico_design_yy_kNm": pilar.momento_resistente_plastico_total_design_yy / 1_000_000,
}

for nome, valor in resumo.items():
    print(f"{nome}: {valor:,.2f}")
```
